```python
# ==========================================
# STEP 1: IMPORT LIBRARIES
# ==========================================
print("Step 1: Importing libraries...")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.datasets import fetch_california_housing
print("Libraries imported successfully!\n")

# Set up plotting style
sns.set_theme(style="whitegrid")

# ==========================================
# TASK 1 — LINEAR REGRESSION (House Prices)
# ==========================================
print("=== STARTING TASK 1: LINEAR REGRESSION ===")

# Load built-in California Housing data
print("Loading California Housing dataset...")
california = fetch_california_housing(as_frame=True)
df_house = california.frame

# Step 3: Explore Dataset
print(f"House Dataset Shape: {df_house.shape}")
print("\n--- First 3 rows of House Dataset ---")
print(df_house.head(3))
print("\nMissing values check:\n", df_house.isnull().sum())

# Step 4: Select Features and Target
X_reg = df_house[['AveRooms', 'AveBedrms', 'Population']]
y_reg = df_house['MedHouseVal']

# Step 5: Split Dataset (80% Train, 20% Test)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)
print(f"Data split done. Training shapes: {X_train_reg.shape}, Testing shapes: {X_test_reg.shape}")

# Step 6: Train Linear Regression Model
print("Training Linear Regression model...")
lin_model = LinearRegression()
lin_model.fit(X_train_reg, y_train_reg)

# Step 7: Make Predictions
print("Making predictions on test data...")
y_pred_reg = lin_model.predict(X_test_reg)

# Step 8: Evaluate the Model (RMSE)
mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)
print(f">>> Task 1 Result - Linear Regression RMSE: {rmse:.4f}\n")


# ==========================================
# TASK 2 — LOGISTIC REGRESSION (Titanic Survival)
# ==========================================
print("=== STARTING TASK 2: LOGISTIC REGRESSION ===")

# Step 1: Load Dataset (Using a reliable online public URL for Titanic data)
print("Loading Titanic dataset from web URL...")
titanic_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df_titanic = pd.read_csv(titanic_url)
print(f"Titanic Dataset Shape: {df_titanic.shape}")

# Step 2: Data Cleaning
print("Cleaning data (handling missing values and encoding text)...")
# Fill missing age values with the median age
df_titanic['Age'] = df_titanic['Age'].fillna(df_titanic['Age'].median())
# Convert categorical variable 'Sex' into numbers (0 for male, 1 for female)
df_titanic['Sex'] = df_titanic['Sex'].map({'male': 0, 'female': 1})

# Step 3: Select Features and Target
X_clf = df_titanic[['Pclass', 'Sex', 'Age', 'Fare']]
y_clf = df_titanic['Survived']

# Step 4: Split Dataset (80% Train, 20% Test)
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42
)

# Step 5: Train Logistic Regression Model
print("Training Logistic Regression model...")
log_model = LogisticRegression()
log_model.fit(X_train_clf, y_train_clf)

# Step 6: Make Predictions
print("Making survival predictions...")
y_pred_clf = log_model.predict(X_test_clf)

# Step 7: Evaluate the Model (Accuracy)
accuracy = accuracy_score(y_test_clf, y_pred_clf)
print(f">>> Task 2 Result - Logistic Regression Accuracy: {accuracy * 100:.2f}%\n")
print("=== ALL TASKS COMPLETED SUCCESSFULLY ===")
```
